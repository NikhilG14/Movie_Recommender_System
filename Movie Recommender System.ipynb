## Performing Item-Based Collaborative Filtering on One Movie Sample

```python
movie_ratings_df
```

```python
userid_movietitle_matrix = movie_ratings_df.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
```

```python
userid_movietitle_matrix
```

```python
titanic = userid_movietitle_matrix['Titanic (1997)']
```

```python
titanic.head()
```

```python
starwars = userid_movietitle_matrix['Star Wars (1977)']
```

```python
starwars.head()
```

### For Titanic

```python
titanic_correlations = pd.DataFrame(userid_movietitle_matrix.corrwith(titanic), columns = ['Correlations'])
```

```python
titanic_correlations = titanic_correlations.join(ratings_mean_count_df['count'])
titanic_correlations.head()
```

```python
# Dropping records with NaN in correlation
titanic_correlations.dropna(inplace = True)
titanic_correlations.head()
```

```python
titanic_correlations.sort_values('Correlations', ascending = False)
```

```python
titanic_correlations[titanic_correlations['count']>80].sort_values('Correlations', ascending = False).head()
```

### For Star Wars

```python
starwars_correlations = pd.DataFrame(userid_movietitle_matrix.corrwith(starwars), columns = ['Correlations'])
```

```python
starwars_correlations = starwars_correlations.join(ratings_mean_count_df['count'])
starwars_correlations.head()
```

```python
# Dropping records with NaN in correlation
starwars_correlations.dropna(inplace = True)
starwars_correlations.head()
```

```python
starwars_correlations.sort_values('Correlations', ascending = False)
```

```python
starwars_correlations[starwars_correlations['count']>80].sort_values('Correlations', ascending = False).head()
```

## Item-based Collaborative Filtering for Entire Dataset

```python
 userid_movietitle_matrix
```

```python
movie_correlations = userid_movietitle_matrix.corr(method = 'pearson', min_periods = 80)
```

```python
movie_correlations
```

```python
myRatings = pd.read_csv('My_Ratings.csv')
myRatings
```

```python
similar_movies_list = pd.Series()

for i in range(2):
    similar_movie = movie_correlations[myRatings['Movie Name'][i]].dropna()
    similar_movie = similar_movie.map(lambda x: x* myRatings['Ratings'][i])
    similar_movies_list = similar_movies_list.append(similar_movie)
```

```python
similar_movies_list.sort_values(inplace = True, ascending = False)
print(similar_movies_list.head(10))
```
